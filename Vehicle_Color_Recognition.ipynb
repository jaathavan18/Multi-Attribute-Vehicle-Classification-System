```python
import os

# Experiment setup
def get_next_exp_folder(base_path):
    os.makedirs(base_path, exist_ok=True)
    num = 0
    while True:
        exp_folder = os.path.join(base_path, str(num))
        if not os.path.exists(exp_folder):
            os.makedirs(exp_folder)
            return exp_folder
        num += 1

expt = "results/col"
exp_folder = get_next_exp_folder(expt)
print(f"Experiment folder: {exp_folder}")

# Data paths
data_dir = 'Data/VCR'
train_dir = os.path.join(data_dir, 'train')
val_dir = os.path.join(data_dir, 'val')
test_dir = os.path.join(data_dir, 'test')
```

```python
from ultralytics import YOLO
import torch
from pathlib import Path

# Classes
train_path = Path(data_dir) / 'train'
classes = [d.name for d in train_path.iterdir() if d.is_dir()]
print(f"Classes: {classes}")
```

```python
model = YOLO('yolo11s-cls.pt')
```

```python
# Train model
results = model.train(
    data=data_dir,
    epochs=50,
    imgsz=224,
    batch=32,
    project=exp_folder,
    name='color_model',
    # plots=False  # Disable plotting
)
```
